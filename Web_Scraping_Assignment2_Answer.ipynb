```python
#Opening the naukri page on automated chrome browser
driver.get("https://www.flipkart.com/")
```

```python
#enetring designation as required in the question

product_name = driver.find_element(By.CLASS_NAME, "_3704LK")
product_name.send_keys('Sunglasses')
```

```python
print(len(brand_name),len(p_description),len(p_price),len(discount))
```
